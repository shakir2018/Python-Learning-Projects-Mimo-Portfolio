```python
"""
ToDo List Manager
A simple command-line task management application.
"""

todo_list = []

while True:
  if len(todo_list) == 0:
    print("Your ToDo lists is empty")

  else:
    print("You have ToDo lists")
    index = 1
    for task in todo_list:
      print(f"{index}. {task}")
      index += 1

  print("Options:")
  print("1. Add Task")
  print("2. Remove Task")
  print("3. Quit")
  choice = input("User's choice:")

  if choice == "1":
    print("Adding task")
    new_task = input("Enter the task: ")
    todo_list.append(new_task)
    print(f"Task {new_task} added")
    
  elif choice == "2":
    if len(todo_list) == 0:
      print("Todo list is empty")

    else:
      todo_list.pop()
    
  elif choice == "3":
    print("Quitting")
    break

  
```

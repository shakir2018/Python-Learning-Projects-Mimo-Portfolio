```python
"""
Restaurant Food Order System
A command-line application for ordering food from multiple cuisines.
"""

italian_food = ["Pasta Bolognese", "Peppero ni pizza", "Margherita pizza","Lasagna"]
indian_food = ["Curry", "Chutney", "Samosa", "Naan"]

def find_meal(name, menu):
    
  """
  Search for a specific meal in a menu dictionary.
    
  Args:
  name (str): Name of the meal to find
  menu_dict (dict): Dictionary of meals and prices
  Returns:
  str or None: The meal name if found, None otherwise
  """  

  if name in menu:
    return name 
  else:
    return None

def select_meal(name, food_type):
    
  """
  Select a meal from a specific cuisine menu.
  Args:
  name (str): Name of the meal
  food_type (str): Type of cuisine
  Returns:
  str or None: The meal name if found, None otherwise
  """
  # Case-insensitive food type matching
  if food_type == "Italian":
    return find_meal(name, italian_food)
  elif food_type == "Indian":
    return find_meal(name, indian_food)
  else:
    return None
  select_meal = find_meal(name, italian_food)
  return select_meal

def display_available_meals(food_type):
    
  """
  Display all available meals for a specific cuisine type.
  Args:
  food_type (str): Type of cuisine
  """
  # Case-insensitive food type matching  
  if food_type == "Italian":
    print("Available Italiadn Meals:")
    for name in italian_food:
      print(name)
  elif food_type == "Indian":
    print("Available Italiadn Indian:")
    for name in indian_food:
      print(name)
  else:
    print("Invalid food type")

def create_summary(name, amount, food_type):
    
  """
  Create an order summary after validating the meal selection.
  Args:
  name (str): Name of the meal ordered
  amount (str): Quantity ordered
  food_type (str): Type of cuisine
  Returns:
  str: Order confirmation message or error message
  """  
  order = select_meal(name, food_type)
  if order:
    # Validate quantity
    return f"You have order {name} {amount}"
  else:
    return f"Meal not found"

print(f"Welcome to the food order Sytem!")
type_input = input("Enter the type of food: ")
display_available_meals(type_input)
name_input = input("Enter meal choice: ")
amount_input = input(input("Enter order quantity: "))
result = create_summary(name_input, amount_input, type_input)
print(result)
```
